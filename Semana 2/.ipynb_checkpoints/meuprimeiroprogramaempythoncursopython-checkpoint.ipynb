## Variáveis
#### Variáveis são um dos grandes pilares da programação em python. Ao invés de trabalhar diretamente com valores, atribuímos valores às variáveis e trabalhamos com elas, o que torna a programação muito mais flexível. Vejamos alguns exemplos de como isto funciona:

```python
# Para atruibuir valores a uma variável, utiliza-se '=':
x = 4
y = 2

# Print é uma função para exibir coisas na tela. Mais para frente você terá uma aula sobre funções!

print(x*y)
print(x+y)
print(y**x)
```

**Nomes de variáveis** \
Um nome de variável deve ser uma sequência de letras e números, e devem sempre começar com uma letra. Não se pode começar o nome com números ou caracteres especiais, com a única excessão sendo o underline: "_":

```python
# Nomes válidos:
num2 = 2
texto = "Olá"
__teste__ = 10
```

```python
# Nomes inválidos:
2num = 2
te%xto = "Olá"
 teste = 10
```

Você pode não ter percebido ainda, mas variáveis são muito úteis! Acostume-se a usá-las, e lembre-se que você pode dar o nome que quiser para elas, então facilite sua compreensão: ao trabalhar com uma variável de idade, chame de "idade", caso seja o tempo: "tempo_em_minutos", por exemplo!

## Operações com texto (string)
#### Strings são as variáveis de texto em Python, marcadas por aspas: " " ou ' ' 


```python

print("Hello World!")
```

```python
# Definindo uma variável para a string
string = "Bem vindos ao curso de Python do FEA.dev!"
#Dando print
print(string)
```

```python
# Podemos também usar uma sequência de três aspas duplas e escrever strings que percorrem várias linhas:
string2 = """
Bem vindos ao curso de Python do FEA.dev!
Esta é a primeira aula
Aproveitem e aprendam muito!
"""
print(string2)
```

#### Agora veremos algumas possíveis operações com elas:

**Concatenamento de strings** \
Em geral, não é possível usar os operadores aritméticos que vimos em strings. Mas é possível 'juntá-los' com o operador '+':

```python
x = "Hello "
y = "World"
print(x + y)
```

**Indexação e fatiamento de strings** \
É possível acessar um caractere específico da sequência utilizando a posição desse caractere, utilizando o que chamamos de índice. Importante notar que, como a maioria das coisas em python, a contagem começa no 0!

```python
# Selecionando o 1 caractere
string = "Bem vindos ao curso de Python do FEA.dev!"
string[0]
```

```python
# Podemos também acessar de trás para frente, sendo o último caractere -1 e assim por diante
# Selecionando o penúltimo caractere
string[-1]
```

```python
# Para selecionar blocos de caracteres, utilizando ':':
# Repare que o começo da seleção inclui o número inicial e exclui o final, ou seja,
# selecionando [4:10] obtemos os caracteres 4 ao 9
string[4:10]
```

```python
# Para selecionar de um ponto até o final ou começo, basta usar o número desejado e dois pontos:
# Pegando do 14 até o final
string[14:]
```

**Métodos de strings** \
As strings oferecem métodos que executam várias operações úteis. Métodos são comandos colocados aoo final da string para realizar determinadas ações e modificações. \
Dentre os principais métodos aplicáveis a strings e suas funcionalidades podemos citar:
<br>

1. **exemplo.upper()**: devolve a string 'exemplo' toda em letras maiúsculas.

2. **exemplo.lower()**: devolve a string 'exemplo' toda em letras minúsculas.

3. **exemplo.strip()**: devolve a string 'exemplo' retirando possíveis espaços em branco no início e no fim da string.

4. **exemplo.startswith('xyz')**: testa se a string 'exemplo' começa com a string 'xyz'.

5. **exemplo.endswith('xyz')**: testa se a string 'exemplo' termina com a string 'xyz'.

6. **exemplo.find('xyz')**: procura a string 'xyz' dentro de 'exemplo' e retorna o primeiro índice onde 'xyz' começa ou retorna -1 se nada for encontrado.

7. **exemplo.replace('old','new')**: retorna uma string nova onde todas as ocorrências de 'old' encontradas em 'exemplo' serão substituídas por 'new'.


Além desses principais, existem outros vários métodos para strings. Esse [link](https://docs.python.org/pt-br/3/library/stdtypes.html#textseq) é um bom ponto de partida para quem quiser conhecer outros exemplos.
Vamos ver aqui as 3 primeiras, mas pode testar as demais!

```python
#Upper:
exemplo = "exemplo"
exemplo.upper()
```

```python
#Lower
exemplo = "ExeMPLo"
exemplo.lower()
```

```python
#Strip
exemplo = "          exemplo     "
exemplo.strip()
```

```python
exemplo2 = "Programar é muito chato"
exemplo2.replace('chato', 'bom')
```

## Input e Output
#### Veremos agora como utilizar informações externas no seu código!

Para perguntar informações para o usuário, utiliza-se a função "input()":

```python
# O string colocado dentro da função será exibido como mensagem para o usuário
input("Como você está? \n")
```

Como já vimos, para exibir mensagens, utilizamos a função "print()":

```python
print("Parabéns por chegar até aqui!")
```

Podemos atribuir inputs à variáveis para trabalhar com estes dados:

```python
idade = input("Quantos anos você tem? \n")

# Duas maneiras de colocar variáveis com strings em prints:

print("Você possui", idade, "anos, que legal!")
print(f'Você possui {idade} anos, que legal!')
```

**f-strings** \
Esta string neste último print que você viu é chamado de f-string. Ao colocar um f na frente, você pode adicionar variáveis, operações, fórmulas e métodos dentro de {} nas strings!

```python
a = 4
b = 2
bebida = "chá"

f_string = f"Tomei incríveis {a*b} xícaras de {bebida.upper() + b*'!'}"

print(f_string)
```

## Extra
##### Apenas algins spoilers do que você logo vai conseguir fazer! Loops e condicionais na próxima semana!

```python
while True:
    intencao = input("Pretende consumir bebidas alcoólicas? Responda com 'Sim' ou 'Não': \n")
    if intencao == 'Sim':
        idade = int(input('Qual sua idade? \n'))
        if idade >= 18:
            print("Sem problemas, pode consumir bebidas alcoolicas!")
            break
        else:
            print('Melhor esperar algum tempo, ainda não posso te fornecer bebidas...')
            continue
    elif intencao == 'Não':
        print('Aproveite, continue assim!')
        break
    else:
        print('Não entendi...')
        continue
```
