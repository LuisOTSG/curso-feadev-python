## Variveis
#### Variveis so um dos grandes pilares da programao em python. Ao invs de trabalhar diretamente com valores, atribumos valores s variveis e trabalhamos com elas, o que torna a programao muito mais flexvel. Vejamos alguns exemplos de como isto funciona:

```python
# Para atruibuir valores a uma varivel, utiliza-se '=':
x = 4
y = 2

# Print  uma funo para exibir coisas na tela. Mais para frente voc ter uma aula sobre funes!

print(x*y)
print(x+y)
print(y**x)
```

```python
# Nomes vlidos:
num2 = 2
texto = "Ol"
__teste__ = 10
```

```python
# Nomes invlidos:
2num = 2
te%xto = "Ol"
 teste = 10
```

#### Agora veremos algumas possveis operaes com elas:

**Concatenamento de strings** \
Em geral, no  possvel usar os operadores aritmticos que vimos em strings. Mas  possvel 'junt-los' com o operador '+':

```python
x = "Hello "
y = "World"
print(x + y)
```

**Indexao e fatiamento de strings** \
 possvel acessar um caractere especfico da sequncia utilizando a posio desse caractere, utilizando o que chamamos de ndice. Importante notar que, como a maioria das coisas em python, a contagem comea no 0!

```python
# Selecionando o 1 caractere
string = "Bem vindos ao curso de Python do FEA.dev!"
string[0]
```

```python
# Podemos tambm acessar de trs para frente, sendo o ltimo caractere -1 e assim por diante
# Selecionando o penltimo caractere
string[-1]
```

```python
# Para selecionar blocos de caracteres, utilizando ':':
# Repare que o comeo da seleo inclui o nmero inicial e exclui o final, ou seja,
# selecionando [4:10] obtemos os caracteres 4 ao 9
string[4:10]
```

```python
# Para selecionar de um ponto at o final ou comeo, basta usar o nmero desejado e dois pontos:
# Pegando do 14 at o final
string[14:]
```

**Mtodos de strings** \
As strings oferecem mtodos que executam vrias operaes teis. Mtodos so comandos colocados aoo final da string para realizar determinadas aes e modificaes. \
Dentre os principais mtodos aplicveis a strings e suas funcionalidades podemos citar:
<br>

1. **exemplo.upper()**: devolve a string 'exemplo' toda em letras maisculas.

2. **exemplo.lower()**: devolve a string 'exemplo' toda em letras minsculas.

3. **exemplo.strip()**: devolve a string 'exemplo' retirando possveis espaos em branco no incio e no fim da string.

4. **exemplo.startswith('xyz')**: testa se a string 'exemplo' comea com a string 'xyz'.

5. **exemplo.endswith('xyz')**: testa se a string 'exemplo' termina com a string 'xyz'.

6. **exemplo.find('xyz')**: procura a string 'xyz' dentro de 'exemplo' e retorna o primeiro ndice onde 'xyz' comea ou retorna -1 se nada for encontrado.

7. **exemplo.replace('old','new')**: retorna uma string nova onde todas as ocorrncias de 'old' encontradas em 'exemplo' sero substitudas por 'new'.


Alm desses principais, existem outros vrios mtodos para strings. Esse [link](https://docs.python.org/pt-br/3/library/stdtypes.html#textseq)  um bom ponto de partida para quem quiser conhecer outros exemplos.
Vamos ver aqui as 3 primeiras, mas pode testar as demais!

```python
#Upper:
exemplo = "exemplo"
exemplo.upper()
```

```python
#Lower
exemplo = "ExeMPLo"
exemplo.lower()
```

```python
#Strip
exemplo = "          exemplo     "
exemplo.strip()
```

```python
exemplo2 = "Programar  muito chato"
exemplo2.replace('chato', 'bom')
```

## Input e Output
#### Veremos agora como utilizar informaes externas no seu cdigo!

Para perguntar informaes para o usurio, utiliza-se a funo "input()":

```python
# O string colocado dentro da funo ser exibido como mensagem para o usurio
input("Como voc est? \n")
```

Como j vimos, para exibir mensagens, utilizamos a funo "print()":

```python
print("Parabns por chegar at aqui!")
```

Podemos atribuir inputs  variveis para trabalhar com estes dados:

```python
idade = input("Quantos anos voc tem? \n")

# Duas maneiras de colocar variveis com strings em prints:

print("Voc possui", idade, "anos, que legal!")
print(f'Voc possui {idade} anos, que legal!')
```

**f-strings** \
Esta string neste ltimo print que voc viu  chamado de f-string. Ao colocar um f na frente, voc pode adicionar variveis, operaes, frmulas e mtodos dentro de {} nas strings!

```python
a = 4
b = 2
bebida = "ch"

f_string = f"Tomei incrveis {a*b} xcaras de {bebida.upper() + b*'!'}"

print(f_string)
```

## Extra
##### Apenas algins spoilers do que voc logo vai conseguir fazer! Loops e condicionais na prxima semana!

```python
while True:
    intencao = input("Pretende consumir bebidas alcolicas? Responda com 'Sim' ou 'No': \n")
    if intencao == 'Sim':
        idade = int(input('Qual sua idade? \n'))
        if idade >= 18:
            print("Sem problemas, pode consumir bebidas alcoolicas!")
            break
        else:
            print('Melhor esperar algum tempo, ainda no posso te fornecer bebidas...')
            continue
    elif intencao == 'No':
        print('Aproveite, continue assim!')
        break
    else:
        print('No entendi...')
        continue
```
