8. Uso do if, elif e else

Essas funções são a forma como as condicionais são expressas/escritas no python. Ao usar o if, você atribui uma condição "if (condição) e depois comandos a serem executados caso a condição seja verdadeira.
O modo de escrever o if no código é assim:

```python
if(condição):
  comando 1
  comando 2
  ...
continuação do código
```

O elif pode ser atribuido como uma continuação do código caso aquela condição não seja verdadeira. Ou seja, você aribui outra condição, que não aquela, e o computador verifica se é verdadeira e executa os comandos desejados. É possível usar mais de 1 elif, seguindo a lógica. A estrutura do elif é assim:

```python
elif(condição):
  comando 1
  comando 2
  ...
continuação do código
```

Já o else, também é opcional, mas diferentemente do elif, ele engloba tudo que não seja as condições antes estabelecidas pelo if e pelo elif e executa os comandos desejados nos casos em que as outras condições não sejam verdadeiras. É possível usar o else só com o if ou com algum/vários elif. A estrutura do else é assim:

```python
else:
  comando 1
  comando 2
  ...
continuação do código
```

É possível pensar no if, elif e else como conjuntos em um espaço amostral. O if é um conjunto, o(s) elif(s) outros, e o else é o conjunto complementar da soma dos conjuntos do if e elif(s) no espaço amostral

Considere as aliquotas e deduções e monte um código para calcular o imposto de renda a depender da renda mensal



```python
# Solicitar ao usuário que insira sua renda mensal
renda_mensal = float(input("Digite sua renda mensal em reais: "))

# Definir as faixas de renda mensal e as alíquotas correspondentes
faixas_de_renda_mensal = [2112.00, 2826.65, 3751.05, 4664.68]
aliquotas_mensais = [0.075, 0.15, 0.225, 0.275]
deducao_por_faixa = [158.40, 370.40, 651.73, 884.96]

# Inicializar o imposto de renda
imposto_de_renda_mensal = 0

# Verificar em qual faixa de renda mensal a pessoa se encontra
if renda_mensal <= faixas_de_renda_mensal[0]:
    imposto_de_renda_mensal = 0
elif renda_mensal <= faixas_de_renda_mensal[1]:
    imposto_de_renda_mensal = renda_mensal * aliquotas_mensais[0] - deducao_por_faixa[0]
elif renda_mensal <= faixas_de_renda_mensal[2]:
    imposto_de_renda_mensal = renda_mensal * aliquotas_mensais[1] - deducao_por_faixa[1]
elif renda_mensal <= faixas_de_renda_mensal[3]:
    imposto_de_renda_mensal = renda_mensal * aliquotas_mensais[2] - deducao_por_faixa[2]
else:
    imposto_de_renda_mensal = renda_mensal * aliquotas_mensais[3] - deducao_por_faixa[3]

# Imprimir o valor mensal do Imposto de Renda
print(f"Seu Imposto de Renda mensal é: R$ {imposto_de_renda_mensal:.2f}")
```

Exemplos com operadores de associação e de identidade

```python
# Exemplo com lista
frutas = ["maçã", "banana", "laranja", "uva"]

if "banana" in frutas:
    print("Banana está na lista de frutas.")

if "morango" not in frutas:
    print("Morango não está na lista de frutas.")

# Exemplo com string
frase = "Python é uma linguagem de programação poderosa."

if "Python" in frase:
    print("A palavra 'Python' está na frase.")

if "Java" not in frase:
    print("A palavra 'Java' não está na frase.")
```

```python
# Criando um dicionário de contatos
contatos = {
    "João": "joao@email.com",
    "Maria": "maria@email.com",
    "Pedro": "pedro@email.com"
}

# Verificando se uma chave está presente no dicionário
nome = "Maria"

if nome in contatos:
    print(f"{nome} está na lista de contatos com o email {contatos[nome]}")
else:
    print(f"{nome} não está na lista de contatos.")

nome = "Carlos"

if nome not in contatos:
    print(f"{nome} não está na lista de contatos.")
```

```python
# Usando operadores de identidade

# Criando duas listas com o mesmo conteúdo
lista1 = [1, 2, 3]
lista2 = [1, 2, 3]

# Verificando se as duas listas têm a mesma identidade
print(lista1 is lista2)
if lista1 is lista2:
    print("lista1 e lista2 têm a mesma identidade.")
else:
    print("lista1 e lista2 têm identidades diferentes.")

# Verificando se as duas listas têm conteúdo igual
print(lista1 == lista2)
if lista1 == lista2:
    print("lista1 e lista2 têm o mesmo conteúdo.")
else:
    print("lista1 e lista2 têm conteúdo diferente.")

# Tornando as 2 a mesma coisa
lista1 = lista2
print(lista1 is lista2)
```
