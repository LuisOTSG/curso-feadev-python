## Bibliotecas Embutidas

Bibliotecas embutidas (ou bibliotecas built-in) referem-se a um conjunto de bibliotecas padro que vm includas na instalao padro do Python. Essas bibliotecas so parte integrante da linguagem Python e esto sempre disponveis para uso, sem a necessidade de instalar pacotes adicionais. Elas fornecem funcionalidades essenciais que podem ser utilizadas em uma ampla variedade de tarefas de programao.

Alguns exemplos de bibliotecas built-in em Python incluem:

### math: 

Fornece funes matemticas para operaes avanadas, como clculos trigonomtricos, exponenciais e logartmicos.

```python
import math
```

- **sqrt(x)**: Esta funo calcula a raiz quadrada de um nmero x.

Exemplo:

```python
math.sqrt(25)
```

- **pow(x, y)**: Calcula x elevado  potncia y.

Exemplo:

```python
math.pow(2, 3)
```

- **floor(x)**: Arredonda um nmero x para baixo, retornando o maior nmero inteiro menor ou igual a x.

Exemplo:

```python
numero = 5.7
math.floor(numero)
```

- **ceil(x)**: Ela retorna o menor nmero inteiro maior ou igual a x

Exemplo:

```python
numero = 5.3
math.ceil(numero)
```

### random: 

Usada para gerao de nmeros aleatrios e amostragem aleatria.

```python
import random
```

- **random()**: Gera um nmero decimal aleatrio entre 0 e 1.

Exemplo:

```python
random.random()
```

- **randint(a, b)**: Gera um nmero inteiro aleatrio entre a e b, inclusive.

Exemplo:

```python
random.randint(1, 10)
```

- **choice(seq)**: Escolhe aleatoriamente um elemento de uma sequncia seq.

Exemplo:

```python
lista = [1, 2, 3, 4, 5, 6, 7]
random.choices(lista)
```

- **shuffle(seq)**: Embaralha aleatoriamente uma sequncia seq (lista, tupla, etc.).

Exemplo:

```python
lista = [1, 2, 3, 4, 5]
random.shuffle(lista)
print(lista)
```

### datetime: 

Manipulao de datas e horas, incluindo formatao e clculos de intervalos de tempo.

```python
import datetime
```

- **datetime.now()**: Retorna o objeto datetime representando a data e hora atuais.

Exemplo:

```python
data_hora_agora = datetime.datetime.now()
print(data_hora_agora)
```

- **date(year, month, day)**: Cria um objeto de data com o ano, ms e dia especificados.

Exemplo:

```python
data = datetime.date(2023, 9, 11)
print(data)
```

- **datetime.strptime(date_string, format)**: Converte uma string em um objeto datetime com base em um formato especfico.

Exemplo:

```python
data_str = "2023-09-10 14:30:00"
data_obj = datetime.datetime.strptime(data_str, "%Y-%m-%d %H:%M:%S")
print(data_obj)
```
